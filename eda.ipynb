## Exploratory Data Analysis (EDA)

EDA Summary:
- The dataset is imbalanced, with more good credit cases than bad credit cases.
- Longer loan duration is associated with higher credit risk.
- Higher credit amount may increase the likelihood of default.
- Age shows a mild relationship with credit risk.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cleaned data (now generated by the previous cells)
df = pd.read_csv("cleaned_data.csv")

# Preview
print("Shape:", df.shape)
display(df.head())

# Target distribution
sns.countplot(x="target", data=df)
plt.title("Distribution of Credit Risk")
plt.xlabel("Credit Risk (0 = Good, 1 = Bad)")
plt.ylabel("Count")
plt.show()

print("Target proportions:")
display(df["target"].value_counts(normalize=True))

# Age vs risk
sns.boxplot(x="target", y="age", data=df)
plt.title("Age vs Credit Risk")
plt.xlabel("Credit Risk (0 = Good, 1 = Bad)")
plt.ylabel("Age")
plt.show()

# Credit amount vs risk
sns.boxplot(x="target", y="credit_amount", data=df)
plt.title("Credit Amount vs Credit Risk")
plt.xlabel("Credit Risk (0 = Good, 1 = Bad)")
plt.ylabel("Credit Amount")
plt.show()

# Loan duration vs risk
sns.boxplot(x="target", y="duration", data=df)
plt.title("Loan Duration vs Credit Risk")
plt.xlabel("Credit Risk (0 = Good, 1 = Bad)")
plt.ylabel("Duration")
plt.show()

# Distribution of numerical variables
df["age"].hist(bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

df["credit_amount"].hist(bins=20)
plt.title("Credit Amount Distribution")
plt.xlabel("Credit Amount")
plt.ylabel("Count")
plt.show()

df["duration"].hist(bins=20)
plt.title("Loan Duration Distribution")
plt.xlabel("Duration")
plt.ylabel("Count")
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap="coolwarm", annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# EDA summary
print("""
EDA Summary:
- The dataset is imbalanced, with more good credit cases than bad credit cases.
- Longer loan duration is associated with higher credit risk.
- Higher credit amount may increase the likelihood of default.
- Age shows a mild relationship with credit risk.
""")
```
