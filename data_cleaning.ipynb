```python
import pandas as pd

# 1. Load data
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/statlog/german/german.data"

columns = [
    "status", "duration", "credit_history", "purpose", "credit_amount",
    "savings", "employment", "installment_rate", "personal_status",
    "other_debtors", "residence_since", "property", "age",
    "other_installment_plans", "housing", "existing_credits",
    "job", "people_liable", "telephone", "foreign_worker", "target"
]

df = pd.read_csv(url, sep=" ", header=None, names=columns)

# 2. Preview data
print("First 5 rows:")
display(df.head())

print("Shape:", df.shape)

print("Data info:")
df.info()

print("Summary statistics:")
display(df.describe())

# 3. Check missing values
print("Missing values:")
display(df.isnull().sum())

# 4. Map target variable
# Original: 1 = good credit, 2 = bad credit
# New: 0 = good credit, 1 = bad credit
df["target"] = df["target"].map({1: 0, 2: 1})

print("Target distribution:")
display(df["target"].value_counts())

# 5. One-hot encode categorical variables
df_encoded = pd.get_dummies(df, drop_first=True)

print("Encoded data shape:", df_encoded.shape)
display(df_encoded.head())

# 6. Save cleaned dataset
df_encoded.to_csv("cleaned_data.csv", index=False)

print("Cleaned data saved as cleaned_data.csv")
```

The dataset was loaded from the UCI German Credit Data source. The target variable was converted into binary format, where 0 represents good credit and 1 represents bad credit. Categorical variables were transformed using one-hot encoding. The cleaned dataset was saved as cleaned_data.csv for later EDA and modeling.
